# 총 인구 수 클러링

```python
# 환경 설정
import requests
from bs4 import BeautifulSoup
```

```python
# HTML 가져오기
url = "https://jumin.mois.go.kr/ageStatMonth.do"

r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
pop_detail = soup.find('div', attrs={'class':'section3'})
print(pop_detail)
```

```python
# 모든 테이블 추출
pop_tr_list = pop_detail.find_all('tr')
print(pop_tr_list)
```

```python
# 태그 안의 텍스트 가져오기

for tr in pop_tr_list:
    if tr.find('td', attrs={'class':'td_admin th1'}):
      print(tr.find('td', attrs={'class':'td_admin th1'}).get_text())
      print(tr.find('td', attrs={'title':'2023년 05월 / 계'}).get_text())
```

```python
# JSON 문자열로 변환
import json

pop_data = {}  # 빈 딕셔너리 생성

for tr in pop_tr_list:
    if tr.find('td', attrs={'class':'td_admin th1'}):
        key = tr.find('td', attrs={'class':'td_admin th1'}).get_text()
        value = tr.find('td', attrs={'title':'2023년 05월 / 계'}).get_text()
        pop_data[key] = value  # 딕셔너리에 키-값 쌍 추가

pop_str = json.dumps(pop_data, ensure_ascii=False)
print(pop_str)
```

```python
# 파이썬 객체로 변환

POP_Data = json.loads(pop_str)

import pandas as pd
POP_Data = pd.DataFrame(POP_Data,index=['0'])
POP_Data.columns = ['전국','서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']
POP_Data
```

```python
# 행 열 교환

POP_Data = POP_Data.transpose()
POP_Data
```

# 전기차 상승률 분석

```python
# 2022년 상반기 전기차 총량 데이터
data_2022_b = Total_data['22년 상반기 전기차 수']

# 2022년 하반기 전기차 총량 데이터
data_2022_a = Total_data['22년 하반기 전기차 수']

# 2023년 상반기 전기차 총량 데이터
data_2023_b = Total_data['23년 상반기 전기차 수']

# 상승률 계산
growth_2022 = ((data_2022_a - data_2022_b) / data_2022_b) * 100

growth_2023 = ((data_2023_b - data_2022_a) / data_2022_a) * 100

# 상승률 출력
print(growth_2022)
```

# 분석결과

대한민국의 전기차 상승률이 전체적으로 떨어졌다는 것은 전기차 시장은 전체적으로 정채되는 추세라는 것을 의미합니다.

# 수소 전기차와 일반 전기차 분석


```python
comparison_data = Total_data[['수소 전기차 수', '23년 상반기 전기차 수']]

plt.figure(figsize=(12, 8))  # 그래프 크기 조정
plt.scatter(comparison_data['수소 전기차 수'], comparison_data['23년 상반기 전기차 수'])
plt.xlabel('수소 전기차 개수')
plt.ylabel('2023년 상반기 전기차 총량')
plt.title('수소 전기차 개수 vs 2023년 상반기 전기차 총량')

# 데이터 포인트에 레이블 추가
for i, txt in enumerate(comparison_data.index):
    plt.annotate(txt, (comparison_data['수소 전기차 수'][i], comparison_data['23년 상반기 전기차 수'][i]))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Total_data['22년 상반기 전기차 수'] = Total_data['22년 상반기 전기차 수'].astype(int)
Total_data['22년 하반기 전기차 수'] = Total_data['22년 하반기 전기차 수'].astype(int)
Total_data['23년 상반기 전기차 수'] = Total_data['23년 상반기 전기차 수'].astype(int)
Total_data['수소 전기차 수'] = Total_data['수소 전기차 수'].astype(int)

logged_electric_cars = np.log(Total_data[['22년 상반기 전기차 수', '22년 하반기 전기차 수', '23년 상반기 전기차 수']])
logged_hydrogen_cars = np.log(Total_data['수소 전기차 수'])
logged_population = np.log(Total_data['23년 대한민국 인구 수'])

plt.figure(figsize=(10, 6))
plt.title('전기차와 인구 수(로그 변환)', fontsize=14)
plt.xlabel('지역')
plt.ylabel('로그 변환된 값')

plt.plot(Total_data.index, logged_electric_cars['22년 상반기 전기차 수'], label='2022년 상반기')
plt.plot(Total_data.index, logged_electric_cars['22년 하반기 전기차 수'], label='2022년 하반기')
plt.plot(Total_data.index, logged_electric_cars['23년 상반기 전기차 수'], label='2023년 상반기')
plt.plot(Total_data.index, logged_hydrogen_cars, label='수소 전기차 수')
plt.plot(Total_data.index, logged_population, label='총 인구수')

plt.legend()

plt.show()
```
